#Plotting

```python
def plot_and_save_graph(dName):

  plotStyles = {
    'markers' : ["*","s","o","^","P","X","d"],
    'color' : ['#d63031', '#273c75','#00cec9', '#6c5ce7',  '#00b894', '#0984e3', '#74b9ff'],
    'lineStyle' : 'solid',
    'markerSize' :  10,
  }

  extension = ['timeCollection', 'memoryCollection']
  plt.figure(figsize=(10*len(extension),5))

  for i in range(len(extension)):
    df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Data_mining/Lab1/Output/'+extension[i]+'_'+dName+'.csv',  sep=",", index_col=['Method'])
    # df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Data Mining Assignments/{}{}.csv'.format(dName, extension[i]), sep=",", index_col=['Method'])  
    # print(df)
    minSupportThreshold = np.array(df.columns, dtype= float)*100
    # print(minSupportThreshold)
    minSupportThreshold = np.round(minSupportThreshold, decimals=2)
    methods = np.unique(df.index.get_level_values('Method'))
    noOfMethods = len(methods)  


    noOfThresholds = len(minSupportThreshold)
    diffBetweenThresholds = abs(minSupportThreshold[0] - minSupportThreshold[1])  
    
    plt.subplot(1, len(extension), i + 1).set_title('Dataset: {}'.format(dName), fontsize=20)
    lines = []
    labels = []
    for j in range(noOfMethods):
      values = np.array(df.loc[methods[j]])
      # print(values)

      if extension[i] == 'timeCollection':
        plt.ylabel('Runtime (sec)', fontsize=20)
      else:
        plt.ylabel('Memory (MB)', fontsize=20)
      plt.xlabel('Minimum Support (%)', fontsize=20)

      plt.xlim(minSupportThreshold[0] + diffBetweenThresholds/2, minSupportThreshold[noOfThresholds-1] - diffBetweenThresholds/2)

      line,  = plt.plot(minSupportThreshold, values, color=plotStyles['color'][j], linestyle=plotStyles['lineStyle'], 
      marker=plotStyles['markers'][j], markersize=plotStyles['markerSize'], label=methods[j])
      labels.append(methods[j])
      lines.append(line)

    plt.legend(lines, labels, prop={'size': 18})      
  plt.show()
```

```python
def save_graph(dName, fileExtension='svg'):
  plotStyles = {
    'markers' : ["*","s","o","^","P","X","d"],
    'color' : ['#d63031', '#273c75','#00cec9', '#6c5ce7',  '#00b894', '#0984e3', '#74b9ff'],
    'lineStyle' : 'solid',
    'markerSize' :  10,
  }

  extension = ['timeCollection', 'memoryCollection']
  plt.figure(figsize=(10,10))

  for i in range(len(extension)):
    df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Data_mining/Lab1/Output/'+extension[i]+'_'+dName+'.csv',  sep=",", index_col=['Method'])
    # print(df)
    minSupportThreshold = np.array(df.columns, dtype= float)*100
    # print(minSupportThreshold)
    minSupportThreshold = np.round(minSupportThreshold, decimals=2)
    noOfThresholds = len(minSupportThreshold)
    diffBetweenThresholds = abs(minSupportThreshold[0] - minSupportThreshold[1])
    # print(diffBetweenThresholds)

    methods = np.unique(df.index.get_level_values('Method'))
    noOfMethods = len(methods)    
    
    plt.clf()
    plt.title('Dataset: {}'.format(dName), fontsize=20)
    lines = []
    labels = []
    for j in range(noOfMethods):
      values = np.array(df.loc[methods[j]])
      # print(values)

      if extension[i] == 'Time':
        plt.ylabel('Runtime (sec)', fontsize=20)
      else:
        plt.ylabel('Memory (MB)', fontsize=20)
      plt.xlabel('Minimum Support (%)', fontsize=20)

      plt.xlim(minSupportThreshold[0] + diffBetweenThresholds/2, minSupportThreshold[noOfThresholds-1] - diffBetweenThresholds/2)

      line,  = plt.plot(minSupportThreshold, values, color=plotStyles['color'][j], linestyle=plotStyles['lineStyle'], 
      marker=plotStyles['markers'][j], markersize=plotStyles['markerSize'], label=methods[j])
      labels.append(methods[j])
      lines.append(line)

    plt.legend(lines, labels, prop={'size': 18})      
    plt.savefig('/content/drive/My Drive/Colab Notebooks/Data_mining/Lab1/Plots/{}{}.{}'.format(dName, extension[i], fileExtension))
```

```python
datasets = ['chess', 'mushroom','retail', 'T10I4D100K', 'kosarak']
for dName in datasets:
  plot_and_save_graph(dName)
  save_graph(dName, 'svg')
```
